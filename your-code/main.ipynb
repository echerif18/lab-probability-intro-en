## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
# YOUR CODE HERE

# p(ball=White) = 10/ (10+12+8)
# p(ball_1=White and ball_2=Black) = 10/ (10+12+8)* 8/9+12+8
# p(ball_1=White and ball_2=Black) = 10/ (10+12+8)* 8/10+12+8
# (10C1)*(12C1)*(8C1)/30C3
```

```python
# YOUR CODE HERE
w = 10
r = 12
b = 8
tot = (w+r+b)

p_b_w = w/tot
p_b_w_b_Norep = (w/tot)*(b/(tot-1))
p_b_w_b_rep = (w/tot)*(b/tot)

p_3diff_color = (math.comb(w, 1) * math.comb(r, 1) * math.comb(b, 1))/math.comb(tot, 3)
```

```python
print(round(p_b_w, 2))
print(round(p_b_w_b_Norep,2))
print(round(p_b_w_b_rep,2))
print(round(p_3diff_color,2))
```

## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

Let define two events: C is a cloudy day and R is Rainny day
* $P(C) = 0.4$
* $P(R\setminus C) = 0.5$
* $P(R\setminus\overline{C}) = 0.2$

$P(R) = P(R \cap C)+ P(R \cap \overline{C})$
    $ = P(R \setminus C) \times P(C)  + P(R \setminus \overline{C}) \times P(\overline{C})$
    $ = P(R \setminus C) \times P(C)  + P(R \setminus \overline{C}) \times (1- P(C))$

```python
# YOUR CODE HERE
## Let define two events:
# C is a cloudy day and R is Rainny day
# P(C) = 0.4
# P(R\C) = 0.5
# P(R\not(C)) = 0.2

# P(R) = P(R and C) + P(R and not(C))
# = P(R\C) * P(C) + P(R\not(C))*P(C)

p_c = 0.4
p_r_c = 0.5
p_r_notc = 0.2

p_r = p_r_c*p_c + p_r_notc*(1-p_c)

print(round(p_r,2))
```
